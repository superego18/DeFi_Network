```python
for i in add_list_:
    r = requests.get(i)
    defi = r.json()
    defi_list_result.append(
        {
            'name' : defi['pageProps']['protocolData']['name'],
            'category' : defi['pageProps']['protocolData']['category'],
            'chain' : defi['pageProps']['protocolData']['chain'],
            'chains' : defi['pageProps']['protocolData']['chains'],
            'currentChainTvls': defi['pageProps']['protocolData']['currentChainTvls']
        }
    )
    
```

```python
len(defi_list_result)
```

```python
df = pd.DataFrame(defi_list_result)
```

```python
df
```

```python
df.to_csv("defi_v2.csv")
```

```python
Token_list = []
for i in df.currentChainTvls:
    Token_list.append(i.keys())
```

```python
len(Token_list)
```

```python
df.insert(5, 'Tokens', Token_list)
```

```python
Token_tvl_list = []

for i in df.currentChainTvls:
    Token_tvl_list.append(sum(list(i.values())))
```
