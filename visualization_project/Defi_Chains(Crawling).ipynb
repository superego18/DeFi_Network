```python
defi_list_result = []
for i in add_list_:
    r = requests.get(i)
    defi = r.json()
    defi_list_result.append(
        {           
        'name' : defi['pageProps']['protocolData']['name'],
        'logo' :  defi['pageProps']['protocolData']['logo'],
        'category' : defi['pageProps']['protocolData']['category'],
        'chain' : defi['pageProps']['protocolData']['chain'],
        'description' : defi['pageProps']['protocolData']['description'],
        'currentChainTvls': defi['pageProps']['protocolData']['currentChainTvls']
         })    
```
